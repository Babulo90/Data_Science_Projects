```python
import pandas as pd
df = pd.read_csv(r"C:\Users\tamal\Documents\Fifa_World_Cup_Project\WorldCupPlayers.csv")
df
```

```python
# Step 1: Handle missing values
# Replace missing positions with 'Unknown'
df['Position'].fillna('Unknown', inplace=True)
# Replace missing events with 'No Event'
df['Event'].fillna('No Event', inplace=True)
```

```python
# Step 2: Fix incorrect shirt numbers
# Replace shirt numbers with 0 as NaN
df['Shirt Number'].replace(0, pd.NA, inplace=True)
missing_values = df.isnull().sum()
```
