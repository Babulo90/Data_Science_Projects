```python
import pandas as pd
df = pd.read_csv(r"C:\Users\tamal\Documents\Fifa_World_Cup_Project\WorldCupMatches.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
# drop rows with critical missing values
df = df.dropna(subset=['Year', 'Home Team Name', 'Away Team Name', 'Home Team Goals', 'Away Team Goals'])
```

```python
# fill missing values for less critical columns
df['Win conditions'] = df['Win conditions'].fillna('None')
```

```python
# Remove duplicates
df = df.drop_duplicates()
```

```python
# Fix data types
# Convert 'Year' to integer
df['Year'] = df['Year'].astype(int)
```

```python
# Ensure goal-related columns are numeric
df['Home Team Goals'] = pd.to_numeric(df['Home Team Goals'], errors='coerce')
df['Away Team Goals'] = pd.to_numeric(df['Away Team Goals'], errors='coerce')
```

```python
# Convert 'Attendance' to numeric (drop missing values)
df['Attendance'] = pd.to_numeric(df['Attendance'], errors='coerce')
```

```python
# Drop unnecessary columns
# Dropping columns that might not be needed for analysis
df = df.drop(columns=['RoundID', 'MatchID', 'Referee', 'Assistant 1', 'Assistant 2'])
```
